```python
import numpy as np
from random import randint
import matplotlib.pyplot as plt

M = 100
p = 0.3
N = 10
tmax = 100
vmax = 5
v0 = 0
t = 0

np.random.seed(1)
arr_x = np.random.randint(M, size = N)
arr_x.sort()
arr_jalur = np.random.randint(2, size = N)
no_mobil = []
for i in range (N):
    no_mobil.append(i)
arr_v = [v0]*N #v awal = 0


arr_x_awal = [] #untuk nampung nilai x awal yang sudah di-random
arr_x_interval = [] #untuk nampung nilai x interval[x80,x90]
arr_bool_x = [] #untuk nampung nilai boolean, apakah x sudah melewati batas M yang ditentukan
arr_rata2 = [] #untuk nampung nilai rata2 mobil kembali ke posisi awalnya
arr_n_rata2 = [] #untuk nampung nilai berapa kali suatu mobil melewati posisi awalnya

for nilai in arr_x:
    arr_x_awal.append(nilai) #untuk memasukan nilai dari arr_x awal yang di-random ke array arr_x_awal
    
for i in range(N):
    arr_rata2.append(0) #untuk membuat default rata2 awal mobil ke posisi awal = 0
    arr_n_rata2.append(0) #untuk membuat list n kali mobil ke posisi awal

for i in range(N):
    arr_bool_x.append(False)
```

```python
import pygame
import time
pygame.init()

win = pygame.display.set_mode((1000,600))
largeText = pygame.font.SysFont('Arial', 15)

letak_x = []
j = 50
for i in range(10):
    letak_x.append(j)
    j+=100

pygame.display.set_caption("Model Lalu Lintas Nagel-Schreckenberg")

white = (255,255,255)

def text_objects(text, font):
    textSurface = font.render(text, True, white)
    return textSurface, textSurface.get_rect()
```

```python
def bubbleSort(arr_x,arr_v,arr_jalur,no_mobil,arr_x_awal,arr_bool_x,arr_rata2,arr_n_rata2):
    n = len(arr_x)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr_x[j] > arr_x[j+1] :
                arr_x[j], arr_x[j+1] = arr_x[j+1], arr_x[j]
                arr_v[j], arr_v[j+1] = arr_v[j+1], arr_v[j]
                arr_jalur[j], arr_jalur[j+1] = arr_jalur[j+1], arr_jalur[j]
                no_mobil[j], no_mobil[j+1] = no_mobil[j+1], no_mobil[j]
                arr_x_awal[j], arr_x_awal[j+1] = arr_x_awal[j+1], arr_x_awal[j]
                arr_bool_x[j], arr_bool_x[j+1] = arr_bool_x[j+1], arr_bool_x[j]
                arr_rata2[j], arr_rata2[j+1] = arr_rata2[j+1], arr_rata2[j]
                arr_n_rata2[j], arr_n_rata2[j+1] = arr_n_rata2[j+1], arr_n_rata2[j]
                
def jarak_mobil_kedepan(i,j):
    jarak = 0
    if(arr_x[i]>arr_x[j]):
        jarak = arr_x[i] - arr_x[j]
    elif(arr_x[i]==arr_x[j]):
        jarak = 0
    else:
        jarak = ((M-1) - arr_x[j]) + arr_x[i] + 1
    return jarak

def jarak_mobil(j):
    ketemu = False
    jarak = 0
    for i in range(j+1,N):
        if(arr_jalur[i]==arr_jalur[j]):
            jarak = jarak_mobil_kedepan(i,j)
            ketemu = True
            break;
    if(not ketemu):
        for i in range(0,j-1):
            if(arr_jalur[i]==arr_jalur[j]):
                jarak = jarak_mobil_kedepan(i,j)
                break;
    return jarak

def halangan_depan(j):
    ketemu = False
    halanganDepan = 0
    for i in range(j+1,N):
        if(arr_jalur[i]!=arr_jalur[j]):
            halanganDepan = jarak_mobil_kedepan(i,j)
            ketemu = True
            break;
    if(not ketemu):
        for i in range(0,j-1):
            if(arr_jalur[i]!=arr_jalur[j]):
                halanganDepan = jarak_mobil_kedepan(i,j)
                break;
    return halanganDepan

def tambah_posisi(seed):
    arr_v[i] = min(arr_v[i],d-1)
    #dengan probabilitas p, nilai v akan -1
    np.random.seed(seed)
    prob = arr_v[i] + (-1 * np.random.choice(np.arange(0, 2), p=[0.7,0.3]))
    arr_v[i] = max(0,prob)
    arr_x[i] = arr_x[i] + arr_v[i]

    #rumus sesuai spesifikasi tubes
    if(arr_x[i]>M-1):
        arr_x[i] = arr_x[i] - (M-1) - 1
        arr_bool_x[i] = True
        
def pindah_jalur():
    arr_x[i] = arr_x[i] + arr_v[i]

    #rumus sesuai spesifikasi tubes
    if(arr_x[i]>M-1):
        arr_x[i] = arr_x[i] - (M-1) - 1
        arr_bool_x[i] = True
        
def tulisan_posisi(i,j):
    #tulisan posisi per mobil
    text = 'Posisi '+str(i)+' : '+str(arr_x[j])
    TextSurf, TextRect = text_objects(text, largeText)
    TextRect.center = (letak_x[j],550)
    win.blit(TextSurf, TextRect)
    pygame.display.update()
    
def text_dimobil(i,j):
    #text 0,1,2,3... di mobil
    if(arr_jalur[j] == 0):
        pygame.draw.rect(win,(255,0,0), (temp,y,w,h))
        win.blit(largeText.render('%d'%i, True, (0,0,0)), (temp, y))
    else:
        pygame.draw.rect(win,(255,0,0), (temp,y+50,w,h))
        win.blit(largeText.render('%d'%i, True, (0,0,0)), (temp, y+50))
    pygame.display.update()

def text_time():
    #text time di atas kiri
    TextSurf, TextRect = text_objects('Time : %d'%t, largeText)
    TextRect.center = (100,50)
    win.blit(TextSurf, TextRect)
    pygame.display.update()
```

```python
def jarak_mobil_kebelakang(i,j):
    jarak = 0
    if(i > j):
        if(arr_x[i] > M-30 and arr_x[j] < M-30):
            jarak = ((M-1) - arr_x[i]) + arr_x[j] + 1
        elif(arr_x[j] > M-30 and arr_x[i] < 30):
            jarak = ((M-1) - arr_x[j]) + arr_x[i] + 1
        elif(arr_x[i] > arr_x[j]):
            jarak = arr_x[i] - arr_x[j]
        else:
            jarak = arr_x[j] - arr_x[i]
    else:
        if(arr_x[j] > M-30 and arr_x[i] < 10):
            jarak = ((M-1) - arr_x[j]) + arr_x[i] + 1
        elif(arr_x[i] > M-30 and arr_x[j] < 10):
            jarak = ((M-1) - arr_x[i]) + arr_x[j] + 1
        elif(arr_x[j] > arr_x[i]):
            jarak = arr_x[j] - arr_x[i]
        else:
            jarak = arr_x[i] - arr_x[j]
    return jarak

def halangan_belakang(j):
    halBel = 0
    ketemu = False
    for i in reversed(range(0,j)):
        if(arr_jalur[i]!=arr_jalur[j]):
            halBel = jarak_mobil_kebelakang(i,j)
            ketemu = True
            break
    if(not ketemu):
        for i in reversed(range(j+1,N)):
            if(arr_jalur[i]!=arr_jalur[j]):
                halBel = jarak_mobil_kebelakang(i,j)
                break
    return halBel

def rata2():
    #untuk mencari rata2 nilai
    if(arr_bool_x[i] == True and arr_x[i]>arr_x_awal[i]):
        arr_bool_x[i] = False
        arr_n_rata2[i] += 1
        arr_rata2[i] = arr_rata2[i]+ (t-arr_rata2[i])

def interval():
    #untuk mencari antara interval[x80,x90]
    if(arr_x[i] <= 90 and arr_x[i]>=80):
        arr_x_interval[t] += 1
```

```python
y = 250
w = 10
h = 50
vel = 5
t = 0

seed = 0
run = True
temp = 0
text_time()
urutkan = False
arr_x_interval.append(0)
for i in range(N):
    d = jarak_mobil(i)
    gapahead = halangan_depan(i) - 1
    gaplback = halangan_belakang(i) - 1
    arr_v[i] = min(arr_v[i]+1,vmax)
    if(arr_v[i] >= d and (gaplback >= vmax or gaplback == -1) and (gapahead >= arr_v[i] or gapahead == -1)):
        if(arr_jalur[i] == 0):
            arr_jalur[i] = 1
        else:
            arr_jalur[i] = 0
        pindah_jalur()
    else:
        tambah_posisi(seed)
        seed += 1
    rata2()
    interval()
        
for i in range(1,N):
    if(arr_x[i] > arr_x[i-1]):
        urutkan = True

if(urutkan):
    bubbleSort(arr_x,arr_v,arr_jalur,no_mobil,arr_x_awal,arr_bool_x,arr_rata2,arr_n_rata2)        


    
for i in range(N):
    #untuk nampilin di aplikasi nanti
    temp = arr_x[i] * 10
    text_dimobil(no_mobil[i],i)
    tulisan_posisi(no_mobil[i],i)

while run:
    pygame.time.delay(100)
    
    for event in pygame.event.get():
        if(event.type == pygame.QUIT):
            run = False
            
    keys = pygame.key.get_pressed()
    
    if keys[pygame.K_RIGHT]:
        t += 1
        arr_x_interval.append(0)
        urutkan = False
        if(t <= tmax-1):
            for i in range(N):
                d = jarak_mobil(i)
                gapahead = halangan_depan(i) - 1
                gaplback = halangan_belakang(i) - 1
                arr_v[i] = min(arr_v[i]+1,vmax)
                if(arr_v[i] >= d and (gaplback >= vmax or gaplback == -1) and (gapahead >= arr_v[i] or gapahead == -1)):
                    if(arr_jalur[i] == 0):
                        arr_jalur[i] = 1
                    else:
                        arr_jalur[i] = 0
                    pindah_jalur()
                else:
                    tambah_posisi(seed)
                    seed += 1
                rata2()
                interval()
            for i in range(1,N):
                if(arr_x[i] > arr_x[i-1]):
                    urutkan = True    
            if(urutkan):
                bubbleSort(arr_x,arr_v,arr_jalur,no_mobil,arr_x_awal,arr_bool_x,arr_rata2,arr_n_rata2)
            
            for i in range(N):
                #untuk nampilin di aplikasi nanti
                temp = arr_x[i] * 10
                text_dimobil(no_mobil[i],i)
                tulisan_posisi(no_mobil[i],i)
                
            text_time()
    win.fill((0,0,0))
pygame.quit()
```
